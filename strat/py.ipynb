```python
df = master_df
df.columns.view()
```

```python
selected_cols = ["timestamp", "CLOSE_UNICORN_HORNS", "CLOSE_ELVEN_WINE", "CLOSE_VAMPIRE_BLOOD", "CLOSE_PHOENIX_FEATHERS"]
df_sel = df[selected_cols]
df_sel
```

```python
# here, lets define the lookback. 

lookback = 500
```

```python
# calculating the % ret


close_cols = [col for col in df_sel.columns if col.startswith("CLOSE_")]
df_rets = pd.DataFrame()
for col in close_cols:
    df_rets[f'Return_{col}'] = ((df_sel[col] - df_sel[col].shift(lookback)) / df_sel[col].shift(lookback)) * 100

df_rets
```

```python
# now, we find the maximum in df_rets, to give a signal for momentum strat.

df_rets["Max_ret"] = df_rets[1:].max(axis=1)
ret_cols = [col for col in df_rets.columns if col.startswith("Return")]
df_rets["Symbol_for_max_ret"] = df_rets[ret_cols].idxmax(axis=1)

df_rets
```
